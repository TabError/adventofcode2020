```python
import re

with open("07bags") as f:
    rules = f.readlines()

rules = [r.replace('.\n', '') for r in rules]

pattern = re.compile(r"(.*?) bags contain (.*)")
rules = [pattern.match(r) for r in rules]

rules = [(r.group(1), r.group(2).split(", ")) for r in rules]

bag_pattern = re.compile(r"\d* (.*?) bag")
rules = [(p, [bag_pattern.match(b) for b in l]) for p, l in rules]
```

```python
rules = [(p, [b.group(1) if b != None else None for b in l]) for p, l in rules]
```

```python
rules
```

```python
parents = []
ours = "shiny gold"
for p, l in rules:
    if ours in l:
        parents.append(p)
parents
```

```python
index = 0
while index < len(parents):
    for bag, contained_bags in rules:
        if parents[index] in contained_bags and bag not in parents:
            parents.append(bag)
    index += 1
    print(parents)
parents
```
